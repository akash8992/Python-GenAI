```python
# 1. What is a function in Python?
def greet(name):
    # This function returns a greeting message
    return f"Hello, {name}!"

print(greet("Akash"))

# 2. Difference between function and method
def add(a, b):
    return a + b  # This is a function

text = "hello"
uppercase_text = text.upper()  # upper() is a method
print(uppercase_text)

# 3. Using *args and **kwargs
def demo(*args, **kwargs):
    print("Positional arguments (args):", args)
    print("Keyword arguments (kwargs):", kwargs)

demo(1, 2, 3, name="Alice", age=30)

# 4. Lambda function (anonymous function)
multiply = lambda x, y: x * y
print("Lambda result:", multiply(4, 5))  # Output: 20

# 5. Difference between return and print
def return_vs_print():
    return "This is return"  # Value is returned, not printed

result = return_vs_print()
print("Returned value:", result)

# 6. Returning multiple values from a function
def operations(a, b):
    return a + b, a - b  # Returns a tuple (sum, difference)

sum_val, diff_val = operations(10, 5)
print("Sum:", sum_val)
print("Difference:", diff_val)

# 7. Recursion (e.g., factorial)
def factorial(n):
    # Recursive base case
    if n == 0:
        return 1
    # Recursive step
    return n * factorial(n - 1)

print("Factorial of 5:", factorial(5))  # Output: 120

# 8. Variable scope inside a function
def scope_example():
    local_var = 100  # local variable
    print("Inside function:", local_var)

scope_example()
# print(local_var)  # Uncommenting this will raise an error: variable not defined outside the function

# 9. Docstring example
def square(n):
    """Returns the square of a number"""
    return n * n

print("Docstring of square():", square.__doc__)

# 10. Higher-order function (takes another function as input)
def shout(text):
    return text.upper()

def greet_hof(func):
    # Function is passed as an argument
    return func("hello from HOF")

print(greet_hof(shout))  # Output: HELLO FROM HOF
```
